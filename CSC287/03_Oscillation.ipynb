```python
# Defube a function from eq(7)
# time t : main variable
# A. omega, delta : parameter

def shm(t, A, omega, delta, x0):
  return A*np.sin(omega*t+delta) +x0
```

```python
from scipy.optimize import curve_fit

A = 25
omega = 2*np.pi / 4
delta = np.pi
x0 = 145

[popt, pcov] = curve_fit( shm, t, y, p0=[ A, omega, delta, x0])


# plotting the fitted model along with the data
plt.xlabel('time (s)')
plt.ylabel('Position-y (unit)')
plt.scatter(t,y)
A = popt[0]
omega = popt[1]
delta = popt[2]
x0 = popt[3]
plt.plot(t, shm(t, A, omega, delta, x0), 'r')
plt.show()



# 
# 
# Omega
# X0

period = 2*np.pi/omega
frequency = 1/period
vmax= omega*A

print(popt)
print('Amplitude is ',A , ' unit')
print('Omega is ',omega, ' radian')
print('Delta is ',delta, ' radian')
print('Equilbrium point is ',x0, ' unit')
print('Period is ',period, ' second')
print('Frequency is ',frequency, ' hz')
print('Vmax is ',vmax, ' unit/sec')
```

Damped Haimonic Function

```python
import math

# Defube a function from eq(12)
# time t : main variable
# A0, omega, delta, x0 : parameter

def dhm(t,A, k, omega, delta, x0):
  # omega = math.sqrt((k/m) - (b*b/(4*m*m)))
  temp1 = A * (np.exp(1) ** (-0.5*k*t))
  return  temp1 * np.sin(omega*t + delta) +x0
```

```python
from scipy.optimize import curve_fit

A = 21
k = 0
omega = np.pi*2
delta = -144
x0 = -142

print('Left Tower')

[popt, pcov] = curve_fit(dhm, L_t, L_x, p0=[A, k, omega, delta, x0])
[A0, k, omega, delta, x0] = popt
print('A0: ', A0)
print('k: ', k)
print('omega: ', omega)
print('delta: ', delta)
print('x0: ', x0)


# plotting the fitted model along with the data
plt.xlabel('time (s)')
plt.ylabel('Position-y (unit)')
plt.scatter(L_t, L_x)
plt.plot(L_t, dhm(L_t, A0, k, omega, delta, x0), 'r')
plt.show()


print('Right Tower')

[popt, pcov] = curve_fit(dhm, R_t, R_x)
[A0, k, omega, delta, x0] = popt
print('A0: ', A0)
print('k: ', k)
print('omega: ', omega)
print('delta: ', delta)
print('x0: ', x0)

# plotting the fitted model along with the data
plt.xlabel('time (s)')
plt.ylabel('Position-y (unit)')
plt.scatter(R_t,R_x)
plt.plot(R_t, dhm(R_t, A0, k, omega, delta, x0), 'r')
plt.show()


print('Pendulum')

[popt, pcov] = curve_fit(dhm, P_t, P_x)
[A0, k, omega, delta, x0] = popt
print('A0: ', A0)
print('k: ', k)
print('omega: ', omega)
print('delta: ', delta)
print('x0: ', x0)

# plotting the fitted model along with the data
plt.xlabel('time (s)')
plt.ylabel('Position-y (unit)')
plt.scatter(P_t,P_x)
plt.plot(P_t, dhm(P_t, A0, k, omega, delta, x0), 'r')
plt.show()
```

## Discussion
1. From the equation 12 (Damped Harmonic Motion), increasing in b results in decreasing in omega. Also, A0 is multiply by e^(-e/2m)t which will decrease over the increasing of b. We can conclude that if b increase the x will be close to 0 faster. On the other hand, if m is increase, it will take more time for the result of equation to reach 0.

2. The value of k in the left tower (without pendulum) is very low compare to the tower on the right (with pendulum). From the result, we can assume that as the value of k goes higher, the motion will stop sonner.

3. The phase of the right tower is ahead of the pendulum on it.

4. Because of the pendulum on the top of the right tower, the type of motion changed from simple harmonic motion to damped harmonic motion. The energy of the tower is lost due to the motion of the pendulum on the top of the tower. Therefore, the tower produce less and less amplitude over time until it stop moving.
